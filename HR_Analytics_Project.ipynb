# Observation

 - We observe that there is a relation between avgworkinghrs , lastevaluation , numberofprojects and empsatisfactorylevel, empleaving.
 - The employee who is not satisfied with the company are leaving.
 - The people who are working on less than `3` projects are leaving due to less work and for there career growth.
 - People who are working on more than `5` projects are leaving due to high work pressure.
 - The people working from `127 hrs to 160 hrs` and for more that `275 hrs` are leaving the company due to less working time and higher working hours.
 - The employees are likely to leave the company after working for `2 to 3` years and if the person is working for more than `6` years is likely to stay in the company.
 - The work accedient does not have any effect on the employees leaving the company
 - The employees are leaving the company because of low wages
 - The people working in `sales`, `technical`, `support` are more likely to leave the comapny
